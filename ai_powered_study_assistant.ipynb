importing the modle


```python
from google import genai
from google.colab import userdata
```

initlazing the client communication


```python
client = genai.Client(api_key=userdata.get("GEMINI_API_KEY"))
print(client)
```

use client object and modle for the text genration all the def the ai assitant and open ai modle


```python
def study_assistant(user_prompt):
    response = client.models.generate_content(
        model="models/gemini-3.5-flash",
        contents=user_prompt
    )
    return response
```

```python
output = study_assistant("explain me the RAG ")
print (output.text)
```
