```python
from google import genai
from google.colab import userdata
```

initlazing the client communication


```python
client = genai.Client(api_key=userdata.get("GEMINI_API_KEY"))
print(client)
```

use client object and modle for the text genration all the def the ai assitant and open ai modle


```python
def study_assistant(question):
    prompt = f"""
You are an expert AI Study Assistant and mentor.

Your goals are:
- Explain complex concepts in a simple and beginner-friendly way.
- Teach using the Feynman Technique.
- Break difficult topics into small, logical steps.
- Give real-world examples and analogies.
- Use bullet points, tables, and ASCII diagrams.
- Explain technical terms before using them.
- Avoid unnecessary jargon.
- Highlight important keywords.
- End with:
  1. Key Takeaways
  2. Common Interview Questions
  3. Real-world Applications
  4. Quick Revision Summary

Here is my question:

{question}
"""

    response = client.models.generate_content(
        model="models/gemini-3.5-flash",
        contents=prompt
    )

    return response
```

```python
output = study_assistant("explain me the RAG ")
print (output.text)
```
